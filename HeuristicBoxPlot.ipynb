# Imports and dataset loading

```python
import numpy as np
import matplotlib.pyplot as plt
import math

heuristicArray = ["Max Gain", "WAdGC", "AsHB Flex", "Loss Comp", "Random Forest"]
osnrResult   = []
rippleResult = []

osnr = open("osnrResults", "r")
osnr = osnr.readlines()[0].replace("[[", "").replace("]]", "").split("], [")

for array in osnr:
    array = array.split(',')
    osnrResult.append(list(map(float, array)))
    
ripple = open("rippleResults", "r")
ripple = ripple.readlines()[0].replace("[[", "").replace("]]", "").split("], [")

for array in ripple:
    array = array.split(',')
    rippleResult.append(list(map(float, array)))


```

# Plots

```python
def plotOSNR():
    fig, axis = plt.subplots(figsize = (10, 5))
    data      = osnrResult

    axis.boxplot(data)
    axis.set_title("OSNR")
    axis.set_xticklabels(np.repeat(heuristicArray, 1))

    plt.show()
    
def plotRipple():
    fig, axis = plt.subplots(figsize = (10, 5))
    data      = rippleResult

    axis.boxplot(data)
    axis.set_title("Ripple")
    axis.set_xticklabels(np.repeat(heuristicArray, 1))

    plt.show()
```

```python
plotOSNR()
```

```python
plotRipple()
```
